```python
# ============================================================
# TUGAS PRAKTIKUM PERTEMUAN 01
# ============================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer

# Tugas 1: array random 20 angka, hitung mean, median, std
print("\n=== Tugas 1: Statistik Array Random ===")
np.random.seed(42) # Agar hasil konsisten
data_acak = np.random.randint(1, 101, 20)
print("Array:", data_acak)
print("Mean  :", round(data_acak.mean(), 2))
print("Median:", round(np.median(data_acak), 2))
print("Std   :", round(data_acak.std(), 2))

```

Tugas 1: Buat array acak 20 angka, hitung mean, median, std

Tujuan : Tujuannya untuk mempraktikkan cara pembuatan data simulasi menggunakan library NumPy dan menghitung nilai dasarnya seperti rata-rata, nilai tengah, dan standar deviasi.

Langkah Kerja :
1. Mengimpor library NumPy yang akan digunakan untuk mengelolah arraynya.

2. Mengunci hasil acak agar tidak berubah-ubah setiap kali dijalankan.

3. Membuat 20 data angka secara acak dengan rentang 1 sampai 100.

4. Menghitung nilai mean, median, dan standar deviasi dari data tersebut.

5. Menampilkan hasil perhitungan kelayar dengan format 2 angka dibelakang koma.

Penjelasan Kodenya,
Disini saya menggunakan beberapa fungsi penting dari library NumPy :
1. np.random.seed(42) : ini fungsinya buat "ngunci" angka acaknya. Jadi walaupun kodenya dijalankan berkali-kali dilaptop saya atau laptop aslab, angkanya bakal tetap sama.
2. np.random.randint(1, 101, 20) : Disini saya nyuruh python buat bikin 20 angka acak yang nilainya antara 1 sampai 100.
3. .mean() : Saya pakai fungsi ini buat nyari nilai rata-rata dari semua angka yang udah dibikin tadi.
4. np.median() : Ini buat nyari nilai yang posisinya benar-benar ditengah kalau angkanya diurutin
5. .std() : Ini singkatan dari Standard Deviation. Saya pakai  buat liat seberapa jauh persebaran angka-angkanya dari nilai rata-rata.
6. round(...,2) : Karena biasanya hasil pembagian itu komanya panjang banget, jadi saya pakai fungsi ini biar cuma diambil 2 angka dibelakang koma biar tampilannya rapi.


Kesimpulan singkat :
Dari hasil eksekusi kode di atas, saya berhasil membuat 20 data acak dan mengetahui karakteristik datanya. Nilai rata-rata dan median yang didapatkan cukup berdekatan, yang artinya data yang di buat oleh fungsi random ini memiliki persebaran yang cukup baik dan tidak terlalu timpang.

```python
# Tugas 2: Buat DataFrame nilai 10 pelajar dan cari 3 nilai tertinggi.
print("\n=== Tugas 2: Top 3 Nilai Mahasiswa ===")
daftar_nama = [
    "Nur", "Jake", "Chika", "Dina", "Anabel",
    "Gabriel", "Fahira", "Harry", "Aqeela", "Bella"
]
nilai_mhs = [85, 78, 92, 65, 88, 70, 95, 82, 80, 89]

df_mhs = pd.DataFrame({
    "nama": daftar_nama,
    "nilai": nilai_mhs
})
print(df_mhs.to_string(index=False))

top3 = df_mhs.nlargest(3, "nilai")
print("\nTop 3 Mahasiswa:")
print(top3.to_string(index=False))
```

Tugas 2: Buat DataFrame nilai 10 pelajar dan cari 3 nilai tertinggi.

Tujuan : Untuk tahu cara mengelola data terstruktur menggunakan library Pandas, khususnya dalam membuat tabel (DataFrame) dan melakukan filter data berdasarkan nilai tertentu.

Langkah Kerja :
1. Membuat list nama dan list nilai mahasiswa sebagai sumber data.
2. MMemggabungkan data tersebut ke dalam sebuat tabel.
3. Menggunakan fungsi filter untuk mencari tiga mahasiswa dengan nilai tertinggi.
4. Menampilkan tabel secara keseluruhan dan tabel khusus untuk 3 besar tanpa kolom.

Penjelasan kodenya,
disini saya pakai Pandas karna lebih bagus dari list biasa dalam mengelola tabel :
1. pd.DataFrame() : Saya gunakan fungsi ini buat bungkus data nama dan data nilai jadi satu tabel yang rapi. nama kolomnya juga saya set langsung disini.
2. nlargest(3, "nilai") : ini fungsinya paling instan untuk cari peringkat. saya tinggal masukkan angka 3 dan kolom mana yang mau dicek, hasilnya langsung keluar otomatis.
3. to_string(index=False) : Saya pakai ini supaya angka urutan bawaan python (0,1,2..) nggak ikut muncul di output, jadi hasilnya beneran seperti daftar nilai papan pengumuman.


Kesimpulan singkat :
Penggunaan pandas sangat memudahkan dalam mencari data (seperti peringkat) dalam sebuah dataset. dibandingkan pakai cara manual, fungsi nlargest lebih cepat dan kodenya ringkas.

```python
# Tugas 3: 2 grafik dari data nilai (Ukuran Kecil)
print("\n=== Tugas 3: Visualisasi Data ===")
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# Grafik 1: Bar Chart
axes[0].bar(df_mhs["nama"], df_mhs["nilai"], color="skyblue", edgecolor="navy")
axes[0].set_title("Perbandingan Nilai")
axes[0].tick_params(axis="x", rotation=45)

# Grafik 2: Histogram
axes[1].hist(df_mhs["nilai"], bins=5, color="lightcoral", edgecolor="black")
axes[1].set_title("Distribusi Nilai")

plt.tight_layout()
plt.show() # Menampilkan langsung di notebook
print("Grafik berhasil ditampilkan.")
```

Tugas 3 : Buat 2 grafik dari data nilai.

Tujuannya untuk mengubah data angka menjadi grafik agar pola datanya lebih mudah dipahami secara langsung

Langkah kerja :
1. siapkan area gambar yang dibagi jadi dua bagian.
2. buat grafik batang (Bar chart) untuk perbandingan nilai individu.
3. buat grafik histogram untuk melihat distribusi nilai kelas.
4. Mengatur tata letak agar grafik rapi dan tidak saling tumpuk.

Penjelasan kode, saya gunakan library Matplotlib buat bikin dashboard sederhananya :
1. plt.subplots(1, 2) : saya buat satu baris yang isinya dua kolom grafik supaya bisa dibandingkan langsung grafik batang dan histogram.
2. axes[0].bar() : ini buat bikin grafik batangnya, saya tambahkan rotation=45 biar namanya miring.
3. axes[1].hist(..., bins=5) : grafik ini buat liat kerumunan nilai. saya bgi jadi 5 kelompok buat tahu rentang nilai mana mahasiswa paling banyak berkumpul.
4. plt.tight_layout() : ini fungsi yang wajib ada agar judul dan labelnya tidak tumpang tindih

Kesimpulan singkat : visualisasi ini memberikan gambaran bahwa nilai mahasiswa variatif. Bar chart untuk lihat performa individu dan histogram untuk lihat standar nilai kelas secara umum.

```python
# Tugas 4: Load dataset load_breast_cancer, tulis 3 wawasan
cancer = load_breast_cancer(as_frame=True)
df_cancer = cancer.frame

print("\n=== Tugas 4: Dataset Breast Cancer ===")
print("Ukuran data:", df_cancer.shape)
print("Target     :", list(cancer.target_names))

print(f"\nInsight 1: Dataset memiliki {df_cancer.shape[0]} sampel dengan {df_cancer.shape[1]-1} fitur.")
print(f"Insight 2: Terdapat {df_cancer['target'].value_counts()[0]} sampel Ganas dan {df_cancer['target'].value_counts()[1]} sampel Jinak.")
print(f"Insight 3: Rata-rata 'mean radius' pada seluruh sampel adalah {round(df_cancer['mean radius'].mean(), 2)}")
```

Tugas 4 : Muat dataset load_breast_cancer(), lalu tulis 3 wawasan singkat.

Tujuannya untuk latihan mengelola dataset medis yang sudah ada di library Scikit-Learn dan hasilkan 3 wawasan singkat.

Penjelasan kode, saya menggunakan data asli dari Scikit-Learn untuk simulasi analisis media :
1. load_breast_cancer(as_frame=True) : as_frame ini penting supaya data medisnya jadi tabel pandas, kalau nggak diaktifin nanti datanya cuma berupa angka-angka yang susah dibaca.
2. .shape : saya pakai ini buat cek berapa banyak pasien dan berapa banyak ciri medis yang dicatat dataset ini.
3. value_counts() : fungsi ini saya gunakan buat ngitung jumlah pasien dimasing-masing kategori (ganas atau jinak)
4. .mean() : digunakan buat cari rata-rata ukuran benjolan (mean radius) dari seluruh data yang ada.

Kesimpulan singkat : Analisis (EDA) ternyata sangat penting sebelum masuk ketahap machine learning. dari data ini saya bisa tahu bahwa jumlah sampel cukup banyak dan kita punya target jelas untuk diprediksi nantinya.
